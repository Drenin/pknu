```python
#바 그래프
import csv
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('jeju.csv', encoding = 'cp949')

f = open('jeju.csv')
data = csv.reader(f)
next(data)


result = []
date = df['date']

for row in data:
    row[1:] = map(int, row[1:]) #row의 1부터 끝까지를 int형으로 변경해라
    result.append(sum(row[1:])) #합산
 

#
#제주 무사증 입국제도는 관광을 목적으로 한 외국인은 비자 없이 제주에 입국해 30일간 합법적으로 체류할 수 있게 하는 제도
#2020년 2월부터 잠정 중단
# print(result)
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize = (10, 5))
plt.title('일자별 제주도 직영관광지 관람객현황')
plt.bar(date, result)
plt.xticks(date, rotation = 90)
plt.show()
```

```python
#바 그래프
import csv
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('jeju.csv', encoding = 'cp949')

f = open('jeju.csv')
data = csv.reader(f)
next(data)


result = []
date = df['date']

for row in data:
    row[1:] = map(int, row[1:]) #row의 1부터 끝까지를 int형으로 변경해라
    result.append(sum(row[1:7])) #합산


result_19 = sum(result[7:19])
result_20 = sum(result[20:32])
#print(result)
print(result_19)
print(result_20)
print((result_20 / result_19) * 100, "%")

data = pd.DataFrame({"19년도" : [result_19], "20년도" : [result_20]})
data.plot(kind = 'bar', stacked = True)
plt.title('전년도대비 관광객수')
plt.show()
```
